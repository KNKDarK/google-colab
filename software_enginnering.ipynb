```python
def temperature_modeling(a, b, c, time):
 temperature = a * time**2 + b * time + c
 return temperature
```

```python

print("Hard-coded variables for weather modeling")
a = 1.0
b = 1.0
c = 1.0
time = 2.5
print(f"Temperature from hardcoded coefficients at time {time} hours:")
print(temperature_modeling(a, b, c, time))
```

```python
def read_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    coefficients = [float(line.strip()) for line in lines]
    return coefficients

# Create coefficients.txt if it doesn't exist
file_name_coefficients = "coefficients.txt"
with open(file_name_coefficients, "w") as f:
    f.write("1.0\n")
    f.write("2.0\n")
    f.write("3.0\n")

print(f"Created '{file_name_coefficients}' for weather modeling coefficients.")

print("\nFile input for weather modeling")
a, b, c = read_from_file("coefficients.txt")
time = 0.4
print(f"Temperature from file coefficients at time {time} hours:")
print(temperature_modeling(a, b, c, time))
```

```python
def solve_coefficients(t1, t2, t3, temp1, temp2, temp3):
    import numpy as np
    A = np.array([
        [t1**2, t1, 1],
        [t2**2, t2, 1],
        [t3**2, t3, 1]
    ])
    B = np.array([temp1, temp2, temp3])
    coeffs = np.linalg.solve(A, B)
    return coeffs


print("\nSolving coefficients from system of equations")
a, b, c = solve_coefficients(0, 5, 10, 5, 6, 8)
time = 0.0
print(f"Temperature from system coefficients at time {time} hours:")
print(temperature_modeling(a, b, c, time))
```

```python

import numpy as np
import matplotlib.pyplot as plt


def quadratic(x, a=-1, b=0, c=10):
    return a * x**2 + b * x + c


stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))
y_positions = quadratic(x_positions)


x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, color="blue", linestyle="--", label="Quadratic Curve")


for i in range(len(stages)):
    if i > 0:

        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            color="blue",
            linewidth=2
        )

        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            color="blue",
            linewidth=2
        )


    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)
    plt.text(x_positions[i], y_positions[i] + 0.3, stages[i], ha="center", fontsize=10)


plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```
