```python
# creating a single input file
file_name_single = "single_input.txt"
with open(file_name_single, "w") as f:
    f.write("3")

print(f"Created '{file_name_single}' for single input example.")
```

```python
#reading a single input file and mutlipy it with 2
try:
  with open("single_input.txt", "r") as f:
    num = int(f.read().strip())
    result = num * 2
    print(f"single input read from file:{num} * 2 = {result}")
except FileNotFoundError:
  print("file not found")
except ValueError:
  print("error in reading or converting the input")
```

```python
# Creating  multiple input of file
file_name_multiple = "multiple_inputs.txt"
with open(file_name_multiple, "w") as f:
    f.write("1\n2\n3")

print(f"Created '{file_name_multiple}' for multiple inputs example.")
```

```python
try:
  with open("multiple_inputs.txt","r") as f:
    num = [int(line.strip()) for line in f.readlines()]
    result = sum(num)
    print(f"multiple input read from file:{num} = {result}")
except FileNotFoundError:
  print("file not found")
except ValueError:
  print("error in reading or converting the input")
```
